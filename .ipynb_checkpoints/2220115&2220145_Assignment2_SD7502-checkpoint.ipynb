```python
#2A Define a Data Loader class and transformation pipeline 
```

```python
import torch
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms
from PIL import Image
import pandas as pd

# Defining the transformation pipeline for preprocessing images
transform = transforms.Compose([
    transforms.Resize((224, 224)),  # To resize images to a resolution of 224x224
    transforms.ToTensor(),  # To convert images to Tensor
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])  # To normalize using ImageNet standards
])

# Modified ButterflyDataset class to handle cases with or without labels
class ButterflyDataset(Dataset):
    def __init__(self, csv_file, root_dir, transform=None, has_labels=True):
        self.data = pd.read_csv(csv_file)
        self.root_dir = root_dir
        self.transform = transform
        self.has_labels = has_labels
        
        if has_labels:
            # Encode the labels into integers
            self.data['label'], self.label_mapping = pd.factorize(self.data['label'])

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        img_name = f"{self.root_dir}/{self.data.iloc[idx, 0]}"
        image = Image.open(img_name).convert("RGB")

        if self.has_labels:
            # Get the label if available
            label = self.data.iloc[idx, 1]  # This will now be an integer
            if self.transform:
                image = self.transform(image)
            return image, label
        else:
            # Return only the image if no labels are available
            if self.transform:
                image = self.transform(image)
            return image

# Initialize Dataset and DataLoader for the training set (has labels)
train_dataset = ButterflyDataset(
    csv_file='F:\\Jordan\\Documents\\Github\\SD7502-Project2-2220115_2220145\\Datasets\\archive\\Training_set.csv',
    root_dir='F:\\Jordan\\Documents\\Github\\SD7502-Project2-2220115_2220145\\Datasets\\archive\\train',
    transform=transform,
    has_labels=True  # Indicates that this dataset has labels
)

train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)

# Initialize Dataset and DataLoader for the testing set (no labels)
test_dataset = ButterflyDataset(
    csv_file='F:\\Jordan\\Documents\\Github\\SD7502-Project2-2220115_2220145\\Datasets\\archive\\Testing_set.csv',
    root_dir='F:\\Jordan\\Documents\\Github\\SD7502-Project2-2220115_2220145\\Datasets\\archive\\test',
    transform=transform,
    has_labels=False  # Indicates that this dataset does not have labels
)

test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False)

print("Data loaders initialized successfully!")
```

```python
#Convolutional Neural Network (CNN)

#CNNs is a deep-learning model which performs best in tasks involving image data. CNNs capture spatial hierarcy in iamges via convolutional layers. 
#These layers apply filters to detect patterns such as textures, edges, and complicated shapes. When using CNNs, images are passed through multiple 
#convolutional layers which are then followed by an activation function, e.g., ReLU (Rectified Linear Unit) along with pooling layers to decrease
#dimensionality. Overall, CNN's are highly efficient for image processing which makes it the best deep learning model for image classification.
```

```python
import torch.nn as nn

class CNNModel(nn.Module):
    def __init__(self, num_classes=10):
        super(CNNModel, self).__init__()
        self.conv1 = nn.Conv2d(3, 16, kernel_size=3, padding=1)
        self.conv2 = nn.Conv2d(16, 32, kernel_size=3, padding=1)
        self.conv3 = nn.Conv2d(32, 64, kernel_size=3, padding=1)
        self.pool = nn.MaxPool2d(kernel_size=2, stride=2)
        self.fc1 = nn.Linear(64 * 28 * 28, 128)
        self.fc2 = nn.Linear(128, num_classes)

    def forward(self, x):
        x = self.pool(F.relu(self.conv1(x)))
        x = self.pool(F.relu(self.conv2(x)))
        x = self.pool(F.relu(self.conv3(x)))
        x = x.view(-1, 64 * 28 * 28)
        x = F.relu(self.fc1(x))
        x = self.fc2(x)
        return x

print("Complete!")
```

```python
#Multilayer Perceptron (MLP)

#An MLP is a fully connected feedforward neural network. It comprimises of an input layer, mulitple hidden layers, and an output layer.
#MLP's treat the input data as a one-dimensional vector, meaning the spatial information present in images are not captured.
#Every neuron in MLP is connected to each neuron in the following layer. Activations such as ReLu are used to bring non-linearity.

#Although MLPs can be used to handle image data, they tend to lack the efficiency and effectiveness which is brought upon CNNs.
#This is due to the fact that they are unable to exploit the spatial structure of the images, leading to lower quality result.
#Despite this, however, MLPs are still a popular choice for comparing things across a variety of domains.
```

```python
class MLPModel(nn.Module):
    def __init__(self, num_classes=10):
        super(MLPModel, self).__init__()
        #Flattening the image (3 channels * 224x224)
        self.fc1 = nn.Linear(3 * 224 * 224, 512)
        self.fc2 = nn.Linear(512, 128)
        self.fc3 = nn.Linear(128, num_classes)


    def forward(self, x):
        #Flattening the input
        x = x.view(-1, 3 * 224 * 224)
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = self.fc3(x)
        return x

print("Complete!")
```

```python
#Recurrent Neural Network (RNN) with LSTM

#An RNN with LSTM units are a neural network architecture which is typically used for sequential data. E.g., natural language,
#time series, or for audio. LTSM networks are designed in a way where they can capture temporal dependencies in data via maintaining
#a memory of past inputs through their cells. When using this in the context of image data, RNNs can be adapted by treating the patches or
#rows of an images as sequences to then be fed in the LSTM to model sequential or spatial relationships.

#Although this approach is not typically used for images, it allows the network to process image patches or pixels as a sequence, leading
#to more interesting insights when compared to results from CNN models for example. Although LSTMs have an extra layer of complexity compared
#to the other models, it's more useful when exploring as to how sequence models perform with visual data.
```

```python
# Fix the RNNModel by replacing LTSM with LSTM
class RNNModel(nn.Module):
    def __init__(self, num_classes=10, input_size=64, hidden_size=128, num_layers=2):
        super(RNNModel, self).__init__()
        self.lstm = nn.LSTM(input_size=input_size, hidden_size=hidden_size, num_layers=num_layers, batch_first=True)
        self.fc = nn.Linear(hidden_size, num_classes)

    def forward(self, x):
        # Adjusting input shape to (batch_size, sequence_length, input_size)
        batch_size = x.size(0)
        x = x.view(batch_size, -1, 64)  # Reshape to (batch_size, sequence_length, input_size)
        lstm_out, _ = self.lstm(x)
        lstm_out = lstm_out[:, -1, :]  # Getting the output of the last time step
        x = self.fc(lstm_out)
        return x
print("Complete!")
```

```python
# Check the output of one batch from the train_loader
batch = next(iter(train_loader))
print(type(batch))
print(batch)


# Adjusted training function to unpack only inputs and labels
def train_model(model, train_loader, val_loader, criterion, optimizer, num_epochs=10):
    train_losses = []
    val_losses = []
    best_val_loss = float("inf") 
    patience = 5
    patience_counter = 0

    for epoch in range(num_epochs):
        # Training phase
        model.train()
        running_loss = 0.0
        for batch in train_loader:
            # Assuming batch is a tuple of (inputs, labels)
            inputs, labels = batch[0], batch[1]

            # Ensure labels are numerical (e.g., integers representing class indices)
            if isinstance(labels[0], str):
                raise ValueError("Labels should be numerical indices, but got strings.")

            # Move data to the correct device (GPU or CPU)
            inputs, labels = inputs.to(device), labels.to(device)

            # Zero the parameter gradients
            optimizer.zero_grad()

            # Forward pass
            outputs = model(inputs)
            loss = criterion(outputs, labels)

            # Backward pass and optimization
            loss.backward()
            optimizer.step()

            running_loss += loss.item() * inputs.size(0)

        epoch_train_loss = running_loss / len(train_loader.dataset)  
        train_losses.append(epoch_train_loss)

        # Validation phase
        model.eval()
        running_val_loss = 0.0  
        with torch.no_grad():
            for batch in val_loader:
                inputs, labels = batch[0], batch[1]

                # Ensure labels are numerical
                if isinstance(labels[0], str):
                    raise ValueError("Labels should be numerical indices, but got strings.")

                inputs, labels = inputs.to(device), labels.to(device)

                outputs = model(inputs)
                loss = criterion(outputs, labels)

                running_val_loss += loss.item() * inputs.size(0)

        epoch_val_loss = running_val_loss / len(val_loader.dataset)  
        val_losses.append(epoch_val_loss)

        print(f"Epoch {epoch + 1}/{num_epochs}, Train Loss: {epoch_train_loss:.4f}, Val Loss: {epoch_val_loss:.4f}")

        # Early stopping
        if epoch_val_loss < best_val_loss:
            best_val_loss = epoch_val_loss
            patience_counter = 0
        else:
            patience_counter += 1

        if patience_counter >= patience:
            print("Early stopping triggered")
            break

    return train_losses, val_losses
```

```python
import torch
import torch.nn.functional as F
import torch.nn as nn
import matplotlib.pyplot as plt

# Determine the number of unique labels in the **full training dataset**
num_classes = full_train_dataset.data['label'].nunique()  # Use full_train_dataset to access data attribute
print(f"Number of classes in the dataset: {num_classes}")

# Define the models with the correct number of classes
model1 = CNNModel(num_classes=num_classes)
model2 = MLPModel(num_classes=num_classes)
model3 = RNNModel(num_classes=num_classes, input_size=64, hidden_size=128, num_layers=2)


# List of models and their names for tracking
models = [model1, model2, model3]
model_names = ["Convolutional Neural Network (CNN)", "Multilayer Perceptron (MLP)", "Recurrent Neural Network (RNN)"]

# Define the learning rates to be tested
learning_rates = [0.001, 0.01, 0.1]

# Assuming `train_loader`, `val_loader`, `criterion`, and `device` are already defined
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
criterion = nn.CrossEntropyLoss()  # Assuming a classification problem

# Function to train and visualize each model separately
def train_and_visualize_models_separately(models, model_names, train_loader, val_loader, criterion, learning_rates, num_epochs=5):
    for model_index, (model, model_name) in enumerate(zip(models, model_names)):
        print(f"\n\nStarting Training for {model_name}")
        
        # Create a new figure for each model's visualization
        plt.figure(figsize=(12, 8))
        
        for lr in learning_rates:
            print(f"\nTraining {model_name} with Learning Rate: {lr}")
            
            # Move the model to the correct device (GPU/CPU)
            model.to(device)
            
            # Initialize optimizer with the current learning rate
            optimizer = torch.optim.Adam(model.parameters(), lr=lr)

            # Train the model using the train_model function provided earlier
            train_losses, val_losses = train_model(model, train_loader, val_loader, criterion, optimizer, num_epochs)
            
            # Plot the training and validation losses for this learning rate
            plt.plot(train_losses, label=f'Train Loss (LR={lr})')
            plt.plot(val_losses, label=f'Val Loss (LR={lr})')
            
            # Print the final training and validation loss for this model and learning rate
            print(f"Final Training Loss for {model_name} with LR={lr}: {train_losses[-1]:.4f}")
            print(f"Final Validation Loss for {model_name} with LR={lr}: {val_losses[-1]:.4f}")

            # Reset the model parameters to re-train for each learning rate
            model.apply(lambda layer: layer.reset_parameters() if hasattr(layer, 'reset_parameters') else None)

        # Set up plot labels and titles for the current model
        plt.xlabel('Epochs')
        plt.ylabel('Loss')
        plt.title(f'{model_name} - Training and Validation Loss with Different Learning Rates')
        plt.legend()
        plt.grid(True)
        plt.show()

# Call the function to train and visualize each model separately
train_and_visualize_models_separately(models, model_names, train_loader, val_loader, criterion, learning_rates, num_epochs=5)
```
